```python
from libs.audio import db_amp_change, write_audio, read_audio
from libs.mpl import simple_figure
from matplotlib import pyplot as plt
```

```python
sampling_frequency, data, max_value = read_audio('piano.wav')
simple_figure(sampling_frequency, data)
plt.ylim([-1.5, 1.5])
```

```python
inc_data = db_amp_change(data, 6)
simple_figure(sampling_frequency, inc_data)
write_audio(sampling_frequency, inc_data, max_value, 'task0211_4_1.wav')
plt.ylim([-1.5, 1.5])
```

```python
dec_data = db_amp_change(data, -6)
simple_figure(sampling_frequency, dec_data)
write_audio(sampling_frequency, dec_data, max_value, 'task0211_4_2.wav')
plt.ylim([-1.5, 1.5])
```
